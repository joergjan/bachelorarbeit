```python
import json
from db import get_database
import pandas as pd
import numpy as np
```

```python
# import authentication data

with open('./infos.json') as f:
    infos = json.load(f)
    onet = infos['onet']
    onetUsername = onet['username']
    onetPassword = onet['password']
    mongodb = infos['mongodb']
    mongoUusername = mongodb['username']
    mongoPpassword = mongodb['password']
    mongoUrl = mongodb['connectionString']
```

```python
dbname = get_database()
collection = dbname["onet"]

documents = collection.find()
df = pd.DataFrame(list(documents))
```

```python
df.head(5)
```

```python
print(df.head(1))
```

```python
# Remove empty cols
print("Total documents: ", len(df))

# Drop columns where all values are null
df = df.dropna(how='all', axis=1)

# Drop rows where all values are null
df = df.dropna(how='all', axis=0)

# Replace any remaining NaN values with 0
df = df.fillna(0)

print("Total documents after dropping na: ", len(df))
```

```python
# Remove columns that are not useful
df = df.drop(['additional_information',"related_occupations","work_values","display",'detailed_work_activities'], axis=1)
```

```python
occupation = pd.DataFrame(list(df["occupation"])) 

# List of columns to drop
cols_to_drop = ["code","updated","sample_of_reported_job_titles","notice","summary_resources","details_resources","custom_resources"]

# Check if each column exists in the DataFrame before dropping it
cols_to_drop = [col for col in cols_to_drop if col in occupation.columns]

if len(cols_to_drop) > 0:
    # Drop the columns
    occupation = occupation.drop(cols_to_drop, axis=1)

    df = pd.concat([df.drop('occupation', axis=1), occupation], axis=1)
```

```python
tasks = pd.DataFrame(list(df["tasks"]))

cols_to_drop = [""]

cols_to_drop = [col for col in cols_to_drop if col in tasks.columns]

if len(cols_to_drop) > 0:
    tasks = tasks.drop(cols_to_drop, axis=1)
    df = pd.concat([df.drop('occupation', axis=1), tasks], axis=1)
```
